```python
import statsmodels.api as sm

X = X_cleaned.iloc[ : , 1:]
Y = X_cleaned.iloc[ : , 0]

X = sm.add_constant(X)

model = sm.OLS(Y, X).fit()

print(model.summary())

significant_vars = model.pvalues[model.pvalues < 0.05].index

significant_vars = significant_vars.drop("const", errors="ignore")

X_significant = X[significant_vars]

X_significant
```

```python
from sklearn.ensemble import HistGradientBoostingRegressor
from sklearn.metrics import mean_squared_error
import numpy as np

def shuffle_and_split_df(X_cleaned, test_ratio, target_col="log_price"):
    shuffled_indices = np.random.permutation(len(X_cleaned))
    test_set_size = int(len(X_cleaned) * test_ratio)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    
    X_train = X_cleaned.iloc[train_indices].drop(target_col, axis=1)
    y_train = X_cleaned.iloc[train_indices][target_col]
    X_test = X_cleaned.iloc[test_indices].drop(target_col, axis=1)
    y_test = X_cleaned.iloc[test_indices][target_col]
    
    return X_train, X_test, y_train, y_test

X_train, X_test, y_train, y_test = shuffle_and_split_df(X_cleaned, 0.2)

model = HistGradientBoostingRegressor(
    max_iter=1000,
    early_stopping=True,
    validation_fraction=0.2,
    scoring='neg_root_mean_squared_error',  # correct scoring for RMSE
    n_iter_no_change=10,
    random_state=42
)

model.fit(X_train, y_train)

# Predict and calculate RMSE
y_pred = model.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print(f"Best Iteration: {model.n_iter_}")
print(f"Test RMSE: {rmse:.2f}")
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, FunctionTransformer, PowerTransformer, OneHotEncoder
from sklearn.feature_selection import VarianceThreshold
from sklearn.pipeline import Pipeline
import numpy as np
import pandas as pd 

class GLMProcessor:
    def __init__(self, skew_threshold=0.75, scaling_method='standard'):
        self.skew_threshold = skew_threshold
        self.scaling_method = scaling_method

        self.scaler = None
        self.var_selector = None
        self.log_tranformations = ['host_tenure','days_between_reviews','n_amenities',
                                   'review_gap_days','distance_to_city_center','beds','accommodates','number_of_reviews']
        self.object_transformers = {
            'review_scores_rating': MinMaxScaler()
        }
        self.lambda_transformers = {
            'beds': lambda X_cleaned: np.sqrt(X_cleaned['beds'].clip(lower=0).fillna(0)),
            'host_response_rate_log': lambda X_cleaned: pd.Series(
                np.log1p(X_cleaned['host_response_rate'].fillna(0).clip(lower=1e-5)),
                name='host_response_rate_log'),
            'room_type_encoded': lambda X: pd.get_dummies(
            X_cleaned['room_type']
            .map({0: "Private room", 1: "Entire home/apt", 2: "Shared room"})
            .fillna("Private room"),
            prefix='room_type',
            drop_first=True,
            dtype=int)
        }
        self.drop_features = ['bed_type', 'host_has_profile_pic']
         

    def log_transform(self, L):
        L_transformed = L.copy()

        for col in self.log_tranformations: 
            L[col] = np.log1p(L[col])
        return L
    
    def other_transform(self, L):
        L_transformed = L.copy()
        transformed_dfs = []

        for col, transformer in self.object_transformers.items():
            transformed = transformer.fit_transform(L_transformed[[col]])
            try:
                encoded_cols = transformer.get_feature_names_out([col])
            except AttributeError:
                encoded_cols = [col]

            df = pd.DataFrame(transformed, columns=encoded_cols, index=L_transformed.index)
            transformed_dfs.append(df)
            # Handle lambda-style full-DataFrame transformations

        for col, func in self.lambda_transformers.items():
            try:
                result = func(L_transformed)
            except Exception as e:
                #print(f"[ERROR] Skipping {col}: {e}")
                continue

            if isinstance(result, pd.Series):
                result = result.to_frame()
                encoded_cols = [col]

            elif isinstance(result, pd.DataFrame):
                encoded_cols = result.columns

            elif isinstance(result, np.ndarray):
                if result.ndim == 1:
                    result = result.reshape(-1, 1)
                    encoded_cols = [col]
                else:
                    encoded_cols = [f"{col}_{i}" for i in range(result.shape[1])]

                result = pd.DataFrame(result, columns=encoded_cols, index=L_transformed.index)

            else:
                continue

            transformed_dfs.append(result)


        final_transformed = pd.concat(transformed_dfs, axis=1)

        return final_transformed
       

    def fit_transform(self, L, y=None):
        L = L.copy()

        L_log = self.log_transform(L)

        L_other = self.other_transform(L_log)

        all_transformed = (
            set(self.log_tranformations)
        |   set(self.object_transformers.keys())
        |   set(self.lambda_transformers.keys())
        |   set(self.drop_features)
        )

        untouched_cols = [col for col in L.columns if col not in all_transformed]
        L_pass_through = L[untouched_cols]

        L_final = pd.concat([L_log[self.log_tranformations], L_other, L_pass_through], axis=1)
        L_final = L_final.loc[:, ~L_final.columns.duplicated()]
        return L_final

    def transform(self, L):
        L = L.copy()

        for col in self.skewed_features:
            L[col] = np.log1p(L[col])
        
        L = L[self.selected_features_]
        L_scaled = self.scaler.transform(L)
        return pd.DataFrame(L_scaled, columns=self.selected_features_, index=L.index)

ml_proc = GLMProcessor(skew_threshold=0.75, scaling_method='minmax')
X_model_ready = ml_proc.fit_transform(X_cleaned)
```
